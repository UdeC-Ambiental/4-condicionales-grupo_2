# Integración y reflexión grupal
## ¿Qué ventajas tiene el uso de funciones vectorizadas sobre el uso de los bucles tradicionales?

Usar **funciones vectorizadas** en lugar de bucles normales en Anaconda o Jupyter tiene muchas ventajas. Estas funciones hacen que el programa sea **más rápido**, porque realizan los cálculos de una forma más eficiente dentro del propio sistema, sin tener que repetir paso por paso como en un bucle.  
Además: Usan **menos memoria**, el código se ve **más corto y fácil de entender** y es**menos probable cometer errores** al trabajar con los datos. 
Por eso, cuando se usan herramientas como **NumPy** o **pandas**, las funciones vectorizadas ayudan a que el trabajo con grandes cantidades de información sea mucho más **rápido y sencillo**.

### ¿Cómo ayudan los condicionales y bucles a automatizar análisis ambientales?

Los **condicionales** y **bucles** son herramientas fundamentales para **automatizar procesos repetitivos** en el análisis de datos ambientales.  
Su uso permite **clasificar y procesar grandes volúmenes de información**, como los datos obtenidos en diferentes tipos de muestras de agua, de forma más **eficiente y rápida**.  
Los **condicionales** (`if`, `elif`, `else`) permiten **tomar decisiones** según los valores de los datos.  
Por ejemplo, se pueden clasificar muestras de agua según su **pH**, determinando si son **ácidas**, **básicas** o **neutras**.  
Por su parte, los **bucles** (`for`, `while`) permiten aplicar esas condiciones a **muchos datos a la vez**, sin necesidad de revisarlos uno por uno, lo que **optimiza el análisis y mejora el rendimiento** del procesamiento.

#### ¿Qué aprendieron del trabajo colaborativo con GitHub Classroom?




Es interesante ver cómo se pueden aplicar códigos que automaticen datos; trabajar en conjunto facilita el trabajo en equipo, permitiéndonos acceder a la información de forma rápida y organizada.  
Además, **GitHub** nos permite compartir fácilmente nuestros propios aportes con los demás integrantes del grupo, lo que mejora la coordinación y evita la duplicación de tareas, ya que podemos ver los avances en tiempo real.  

A través de esta experiencia aprendimos a usar **GitHub** como una herramienta colaborativa fundamental, que no solo sirve para guardar código, sino también para mantener un control de versiones, documentar el progreso y aprender de las contribuciones de todo el equipo.  
Esto hace que el trabajo sea más ordenado, profesional y eficiente.
