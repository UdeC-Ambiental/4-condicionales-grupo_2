# üß© Reto 2 ‚Äì pH
**Objetivo:** clasificar pH y generar alerta qu√≠mica.

**Datos:** `../data/muestras_calidad_agua.csv`

### Pasos
1. Carga el CSV en `df`.
2. Extrae `sitio` y `pH`.
3. Crea `clasificacion_pH` con tres rangos.
4. Crea `alerta_quimica` (revisi√≥n si pH < 6.5 o pH > 8.5).
5. Filtra `df_alerta` y exporta `resultado_ph.csv`.

> **Reto adicional:** promedio y desviaci√≥n est√°ndar.

```python
# Escribe tu c√≥digo aqu√≠
import pandas as pd 
# paso 1 : Cargar el archivo CSV 
df = pd.read_csv("muestras_calidad_agua.csv")
# mostar las primeras filas 
df.head()
```

```python
df_ph = df[["sitio", "pH"]]
df_ph
```

```python
def clasificar_ph(valor):
    if valor < 6.5:
        return "√Åcido"
    elif valor <= 8.5:
        return "Neutro"
    else:
        return "B√°sico"

df_ph.loc[:, "clasificacion_pH"] = df_ph["pH"].apply(clasificar_ph)
df_ph
```

```python
df_ph["alerta_quimica"] = df_ph["pH"].apply(lambda x: "Alerta" if (x < 6.5 or x > 8.5) else "Normal")
df_ph
```

```python
# Filtrar las muestras con alerta
df_alerta = df_ph[df_ph["alerta_quimica"] == "Revisar"]

# Mostrar el DataFrame filtrado
df_alerta
```

```python
# Exportar el resultado a un CSV
df_alerta.to_csv("resultado_pH.csv", index=False)
```

```python
import pandas as pd
df = pd.read_csv("muestras_calidad_agua.csv")
df.head()
```

```python
# Extraer solo sitio y pH
df_ph = df[["sitio", "pH"]].copy()
df_ph.head()
```

```python
# Definir funci√≥n de clasificaci√≥n
def clasificar_ph(valor):
    if valor < 6.5:
        return "√Åcido"
    elif valor <= 8.5:
        return "Neutro"
    else:
        return "B√°sico"

# Asignar columna usando .loc para evitar SettingWithCopyWarning
df_ph.loc[:, "clasificacion_pH"] = df_ph["pH"].apply(clasificar_ph)
df_ph.head()
```

```python
# Crear columna de alerta qu√≠mica
df_ph.loc[:, "alerta_quimica"] = df_ph["pH"].apply(lambda x: "Revisar" if (x < 6.5 or x > 8.5) else "Normal")

# Filtrar las muestras con alerta
df_alerta = df_ph[df_ph["alerta_quimica"] == "Revisar"]

# Mostrar las alertas (si hay)
display(df_alerta)

# Exportar a CSV solo si hay filas
if not df_alerta.empty:
    df_alerta.to_csv("resultado_ph.csv", index=False)
    print("‚úÖ Archivo 'resultado_ph.csv' guardado correctamente.")
else:
    print("‚ÑπÔ∏è No hay filas en alerta (df_alerta est√° vac√≠o). No se cre√≥ 'resultado_ph.csv'.")
```
